```python
import pandas as pd
import requests
import pycountry
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## **Dataset Collection**

```python
class url_to_df():
  def __init__(self,url):
    self.url=url.strip()

  def read_url(self):
    response=requests.get(self.url)
    raw_data=response.json()
    return raw_data['value']

  def dataframe(self,age_group):
    data=self.read_url()
    df=pd.DataFrame(data)
    df['Age_group']= str(age_group)
    return df
```

```python

url = "https://ghoapi.azureedge.net/api/NCD_BMI_30C"
age_format='Adult'

read=url_to_df(url)
ob_a=read.dataframe(age_format)
ob_a.head(2)
```

## **Standardize values**

```python
obesity['Gender'] = obesity['Gender'].replace({
    'SEX_MLE': 'Male',
    'SEX_FMLE': 'Female',
    'SEX_BTSX': 'Both'
})

malnutrition['Gender'] = malnutrition['Gender'].replace({
    'SEX_MLE': 'Male',
    'SEX_FMLE': 'Female',
    'SEX_BTSX': 'Both'
})
```

```python
extra_dict = {
    "AFR": "Africa Region",
    "AMR": "Region of the Americas",
    "EMR": "Eastern Mediterranean Region",
    "EUR": "European Region",
    "SEAR": "South-East Asia Region",
    "WPR": "Western Pacific Region",
    "GLOBAL": "Global",

    "WB_LI": "World Bank Low Income",
    "WB_LMI": "World Bank Lower Middle Income",
    "WB_HI": "World Bank High Income",
    "WB_UMI": "World Bank Upper Middle Income"
}

special_cases = {
                    'GLOBAL': 'Global',
                    'WB_LMI': 'Low & Middle Income',
                    'WB_HI': 'High Income',
                    'WB_LI': 'Low Income',
                    'EMR': 'Eastern Mediterranean Region',
                    'EUR': 'Europe',
                    'AFR': 'Africa',
                    'SEAR': 'South-East Asia Region',
                    'WPR': 'Western Pacific Region',
                    'AMR': 'Americas Region',
                    'WB_UMI': 'Upper Middle Income'}

def iso_to_country(code):

    # First check special dictionary
    if code in extra_dict:
        return extra_dict[code]

    if code in special_cases:
        return special_cases[code]

    # Check ISO-3
    country = pycountry.countries.get(alpha_3=code)
    if country:
        return country.name

    # Check ISO-2
    country = pycountry.countries.get(alpha_2=code)
    if country:
        return country.name

    # If nothing matches
    return code
```

```python
obesity['Country'] = obesity['Country'].apply(iso_to_country)

malnutrition['Country'] = malnutrition['Country'].apply(iso_to_country)
```

## **Column create**

```python
obesity['CI_Width']=obesity['HigherBound']-obesity['LowerBound']
print(display(obesity.head(2)))

malnutrition['CI_Width']=malnutrition['HigherBound'] - malnutrition['LowerBound']
print(display(malnutrition.head(2)))
```

```python
def obese(value):
    if value >= 30:
        return 'High'
    elif value >= 25 and value <30 :
        return 'Moderate'
    else:
        return 'Low'

obesity['Obesity_level'] = obesity['MeanEstimate'].apply(obese)
print(display(obesity.head(2)))

def nutrition(value):
    if value >= 20:
        return 'High'
    elif value >= 10 and  value < 20:
        return 'Moderate'
    else:
        return 'Low'

malnutrition['Malnutrition_level'] = malnutrition['MeanEstimate'].apply(nutrition)
print(display(malnutrition.head(2)))
```

```python
import sqlite3

db='nutrition_paradox.db'

try:
  conn=sqlite3.connect(db)
  cursor=conn.cursor()
  print(f'{db} db created ')
except Exception as e:
  print(f'failed in db creation{e}')
```

```python
class storing_db():
  def __init__(self,db,table_name):
    self.db=db
    self.table_name=table_name

  def insert(self,df):
    try:
      df.to_sql(name=self.table_name,con=self.db,if_exists='replace',index=False)
      return f'Successfully inserted {self.table_name} table'

    except Exception as e:
      return f'--- {e} ---'
```

```python
table_name='obesity'

db_obj = storing_db(db=conn,table_name=table_name)
db_obj.insert(obesity)
```

```python
table_name='malnutrition'

db_obj = storing_db(db=conn,table_name=table_name)
db_obj.insert(malnutrition)
```

```python
# Top 5 Regions with Highest Avg Obesity in 2022

pd.read_sql('''SELECT Region,
       AVG(MeanEstimate) AS avg_obesity_2022
FROM obesity
WHERE Year = 2022
GROUP BY Region
ORDER BY avg_obesity_2022 DESC
LIMIT 5''',conn)
```

```python
# Top 5 Countries with Highest Obesity Estimates (Overall)

pd.read_sql('''SELECT Country,
       AVG(MeanEstimate) AS avg_obesity
FROM obesity
GROUP BY Country
ORDER BY avg_obesity DESC
LIMIT 5''',conn)
```

```python
# Obesity Trend in India Over Years

pd.read_sql('''SELECT Year,
       AVG(MeanEstimate) AS india_avg_obesity
FROM obesity
WHERE Country = 'India'
GROUP BY Year
ORDER BY Year''',conn)
```

```python
# Average Obesity by Gender

pd.read_sql('''SELECT Gender,
       AVG(MeanEstimate) AS avg_obesity
FROM obesity
GROUP BY Gender
ORDER BY avg_obesity DESC''',conn)
```

```python
# Country Count by Obesity Level & Age Group

pd.read_sql('''SELECT Obesity_level,
       Age_group,
       COUNT(DISTINCT Country) AS country_count
FROM obesity
GROUP BY Obesity_level, Age_group
ORDER BY Obesity_level, Age_group''',conn)
```

```python
# Least Reliable (Highest Avg CI Width)

pd.read_sql('''SELECT Country,
       AVG(CI_Width) AS avg_ci_width
FROM obesity
GROUP BY Country
ORDER BY avg_ci_width DESC
LIMIT 5''',conn)
```

```python
# Most Consistent (Lowest Avg CI Width)

pd.read_sql('''SELECT Country,
       AVG(CI_Width) AS avg_ci_width
FROM obesity
GROUP BY Country
ORDER BY avg_ci_width ASC
LIMIT 5''',conn)
```

```python
# Average Obesity by Age Group

pd.read_sql('''SELECT Age_group,
       AVG(MeanEstimate) AS avg_obesity
FROM obesity
GROUP BY Age_group
ORDER BY avg_obesity DESC''',conn)
```

```python
# Top 10 Countries with Consistent Low Obesity

pd.read_sql('''SELECT Country,
       AVG(MeanEstimate) AS avg_obesity,
       AVG(CI_Width) AS avg_ci_width
FROM obesity
GROUP BY Country
HAVING AVG(MeanEstimate) < (
        SELECT AVG(MeanEstimate) FROM obesity
      )
   AND AVG(CI_Width) < (
        SELECT AVG(CI_Width) FROM obesity
      )
ORDER BY avg_obesity ASC
LIMIT 10''',conn)
```

```python
# Countries Where Female Obesity > Male (Same Year)

pd.read_sql('''SELECT f.Country,
       f.Year,
       f.MeanEstimate AS female_obesity,
       m.MeanEstimate AS male_obesity,
       (f.MeanEstimate - m.MeanEstimate) AS difference
FROM obesity f
JOIN obesity m
     ON f.Country = m.Country
     AND f.Year = m.Year
WHERE f.Gender = 'Female'
  AND m.Gender = 'Male'
  AND f.MeanEstimate > m.MeanEstimate
ORDER BY difference DESC''',conn)
```

```python
# Global Average Obesity Per Year

pd.read_sql('''SELECT Year,
       AVG(MeanEstimate) AS global_avg_obesity
FROM obesity
GROUP BY Year
ORDER BY Year''',conn)
```
